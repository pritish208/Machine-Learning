```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\prabh\Desktop\data set\Iris\Iris.csv")
df
```

```python
df=df[['PetalLengthCm','PetalWidthCm']]
df
```

```python
x=df['PetalLengthCm']
y=df['PetalWidthCm']
x
```

```python
y
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=0)
```

```python
x=np.array(x_train).reshape(-1,1)
x
```

```python
y=np.array(y_train).reshape(-1,1)
y
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
c=model.intercept_
c
```

```python
m=model.coef_
m
```

```python
y_train_pred1=model.predict(x_train)
y_train_pred1
```

```python
y_pred_test1=model.predict(x_test)
y_pred_test1
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred_test1,color="red")
plt.ylabel('petel Width')
plt.xlabel('petal Length')
plt.title('Training Set')


```
